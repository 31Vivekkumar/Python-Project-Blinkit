```python
print(df['Item Fat Content'].unique())
```

```python
df['Item Fat Content'] = df['Item Fat Content'].replace({'LF': 'Low Fat',
                                                        'low fat': 'Low Fat',
                                                         'reg': 'Regular', })
```

```python
df['Item Fat Content'].unique()
```

```python
total_sales = df['Sales'].sum().round(0)

avg_sales = df['Sales'].mean().__round__()

no_of_items_sold = df['Sales'].count()

avr_rating = df['Rating'].mean().__round__()

print("Total Sales: ", total_sales)
print("Average Sales: ", avg_sales)
print("Number of Items Sold: ", no_of_items_sold)
print("Average Rating: ", avr_rating)
```

```python
sales_by_fat = df.groupby('Item Fat Content')['Sales'].sum()
plt.pie(sales_by_fat, labels=sales_by_fat.index, autopct='%1.1f%%', startangle=140, shadow=True, explode=(0.1, 0)) 
plt.title('Sales by Item Fat Content')
plt.axis('equal')  
plt.show()# Equal aspect ratio ensures that pie is drawn as a circle.
```

TOTAL Sales by Item Type

```python
sales_by_type = df.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))

bars = plt.bar(sales_by_type.index, sales_by_type.values, color='blue')

plt.xticks(rotation=90)
plt.xlabel('Item Type')
plt.ylabel('Sales')
plt.title('Sales by Item Type')

```
